In this chapter, we use a stationary Markov process to construct a process that displays stochastic arithmetic growth, then show how to extract a linear time trend and a martingale. Eventually, we will explore the implications exponentiating this process to transform an arithmetically growing processes like those described in this chapter to construct a process that displays geometric growth.

# 4.1 Definition of additive functional

Let $\left\{W_{t+1}: t \geq 0\right\}$ be a $k$-dimensional stochastic process of unanticipated economic shocks. Let $\left\{X_t: t \geq 0\right\}$ be a discrete-time stationary Markov process that is generated by initial distribution $Q$ for $X_0$ and transition equation
$$
X_{t+1}=\varphi\left(X_t, W_{t+1}\right),
$$
where $\varphi$ is a Borel measurable function. Let $\left\{\mathfrak{A}_t: t=0,1, \ldots\right\}$ be the filtration generated by histories of $W$ and $X ; \mathfrak{A}_t$ serves as the information set (sigma algebra) generated by $X_0, W_1, \ldots, W_t$. We presume that the conditional probability distribution for $W_{t+1}$ conditioned on $\mathfrak{A}_t$ depends only on $X_t$. To assure that the process $\left\{W_{t+1}: t \geq 0\right\}$ represents unanticipated shocks, we restrict it to satisfy
$$
E\left(W_{t+1} \mid X_t\right)=0 .
$$

We condition on a statistical model in the sense of section 1.7 and assume that the stationary $X_t$ process is ergodic.] The Markov structure of $\left\{X_t\right.$ : $t \geq 0\}$ makes the distribution of $\left(X_{t+1}, W_{t+1}\right)$ conditioned on $\mathfrak{A}_t$ depend only on $X_t$.2

Definition 4.1.1. A process $\left\{Y_t\right\}$ is said to be an additive functional if it can be represented as
$$
Y_{t+1}-Y_t=\kappa\left(X_t, W_{t+1}\right)
$$
for a (Borel measurable) function $\kappa: \mathbb{R}^n \times \mathbb{R}^k \rightarrow \mathbb{R}$, or equivalently
$$
Y_t=Y_0+\sum_{j=1}^t \kappa\left(X_{j-1}, W_j\right),
$$
where we initialize $Y_0$ at some arbitrary (Borel measurable) function of $X_0$.
When $Y_0$ is a function of $X_0$, we can construct $Y_t$ as a function of the underlying Markov process between dates zero and $t$.

Definition 4.1.2. An additive functional $\left\{Y_t: t=0,1, \ldots\right\}$ is said to be an additive martingale if $E\left[\kappa\left(X_t, W_{t+1}\right) \mid X_t\right]=0$.

Example 4.1.3. (Stochastic Volatility) Suppose that
$$
\begin{aligned}
Y_{t+1}-Y_t & =\mu\left(X_t\right)+\sigma\left(X_t\right) W_{t+1} \\
X_{t+1} & =\mathbb{A} X_t+\mathbb{B} W_{t+1}
\end{aligned}
$$
where $\left\{W_{t+1}: t \geq 0\right\}$ is an i.i.d. sequence of standardized multivariate normally distributed random vectors, $\mathbb{A}$ is a stable matrix, and $\mathbb{B}$ has full column rank, and the random vector $X_0$ is generated by initial distribution $Q$ associated with the stationary distribution for the $\left\{X_t\right\}$ process. Here $\mu\left(X_t\right)$ is the conditional mean of $Y_{t+1}-Y_t$ and $\left|\sigma\left(X_t\right)\right|^2$ is its conditional variance. This is called a stochastic volatility model because $\left|\sigma\left(X_t\right)\right|^2$ is a stochastic process.

In example (4.1.3), when the conditional mean $\mu\left(X_t\right)=0$, the process $\left\{Y_t\right\}$ is a martingale. Note that $E\left[\kappa\left(X_t, W_{t+1}\right) \mid X_t\right]=0$ implies the usual martingale restriction
$$
E\left(Y_{t+1} \mid \mathfrak{A}_t\right)=Y_t, \text { for } t \geq 0 .
$$
