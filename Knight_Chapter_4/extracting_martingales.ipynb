# 4.2 Extracting Martingales

We can decompose an additive functional into a sum of components, one of which is an additive martingale that encapsulates all long-run stochastic variation as in Proposition 2.2.2. In this section, we show how to extract the martingale component. We adopt a construction like that used to establish Proposition 2.2 .2 and proceed in four steps.
i) Construct the trend coefficient is the unconditional expectation:
$$
\nu=E\left[\kappa\left(X_t, W_{t+1}\right)\right] .
$$
ii) Form the random variable $H_t$ by computing multiperiod forecasts for each horizon and summing these forecasts over all horizons. Start by constructing
$$
\bar{\kappa}(x)=E\left[\kappa\left(X_t, W_{t+1}\right)-\nu \mid X_t=x\right]
$$
Thus
$$
E\left[\kappa\left(X_{t+j-1}, W_{t+j}\right)-\nu \mid X_t=x\right]=\mathbb{T}^{j-1} \bar{\kappa}(x) .
$$

Summing the terms, construct
$$
\begin{aligned}
H_t & =\sum_{j=0}^{\infty} E\left(\left[\kappa\left(X_{t-1+j}, W_{t+j}-\nu\right] \mid X_t\right)\right. \\
& =\kappa\left(X_{t-1}, W_t\right)-\nu+\sum_{j=0}^{\infty} E\left[\bar{\kappa}\left(X_{t+j}\right) \mid X_t\right] \\
& =\kappa_h\left(X_{t-1}, W_t\right)
\end{aligned}
$$
where
$$
\begin{aligned}
\kappa_h\left(X_{t-1}, W_t\right) & =\kappa\left(X_{t-1}, W_t\right)-\nu+\sum_{j=0}^{\infty} \mathbb{T}^j \bar{\kappa}\left(X_t\right) \\
& =\kappa\left(X_{t-1}, W_t\right)-\nu+(\mathbb{I}-\mathbb{T})^{-1} \bar{\kappa}\left(X_t\right)
\end{aligned}
$$
where $\mathbb{T}$ is the operator defined in (3.1). The right side becomes a function of only $\left(X_{t-1}, W_t\right)$ once we substitute for $\varphi\left(X_{t-1}, W_t\right)$ for $X_t$ as implied by (4.1).
This construction requires that the infinite sum
$$
\sum_{j=0}^{\infty} \mathbb{T}^j \bar{\kappa}(x)=(\mathbb{I}-\mathbb{T})^{-1} \bar{\kappa}(x)
$$
converges in mean square relative to the stationary distribution for $\left\{X_t: t \geq 0\right\}$. A sufficient condition for this is that $\mathbb{T}^m$ is a strong contraction for some integer $m \geqslant 1$ and $\bar{\kappa} \in \mathcal{N}$ where $\mathcal{N}$ is defined in (3.9).
iii) Compute
$$
H_t^{+}=E\left(H_{t+1} \mid X_t\right)=\kappa_{+}\left(X_t\right)
$$
where ${ }^3$
$$
\begin{aligned}
\kappa_{+}(x) & \doteq E\left[\kappa\left(X_t, W_{t+1}\right) \mid X_t=x\right]-\nu+E\left[(\mathbb{I}-\mathbb{T})^{-1} \bar{\kappa}\left(X_{t+1}\right) \mid X_t=x\right] \\
& =E\left[\kappa\left(X_t, W_{t+1}\right) \mid X_t=x\right]-\nu+(\mathbb{I}-\mathbb{T})^{-1} \mathbb{T} \bar{\kappa}(x) .
\end{aligned}
$$

iv) Build the martingale increment:
$$
G_t=H_t-H_{t-1}^{+}=\kappa_m\left(X_{t-1}, W_t\right)
$$
where
$$
\kappa_m\left(X_{t-1}, W_t\right)=\kappa_h\left(X_{t-1}, W_t\right)-\kappa_{+}\left(X_{t-1}\right) .
$$

By construction, the expectation of $\kappa_m\left(X_t, W_{t+1}\right)$ conditioned on $X_t$ is zero.

Armed with these calculations, we now report a Markov counterpart to Proposition 2.2.2.

Proposition 4.2.1. Suppose that $\left\{Y_t: t \geq\right\}$ is an additive functional, that $\mathbb{T}^m$ is a strong contraction on $\mathcal{N}$ for some $m$, and that $E\left[\kappa\left(X_t, W_{t+1}\right)^2\right]<$ $\infty$. Then
$$
Y_t=\underbrace{t \nu}_{\text {trend }}+\underbrace{\sum_{j=1}^t \kappa_m\left(X_{j-1}, W_j\right)}_{\text {martingale }}-\underbrace{\kappa_{+}\left(X_t\right)}_{\text {stationary }}+\underbrace{Y_0+\kappa_{+}\left(X_0\right)}_{\text {invariant }} .
$$

Notice that the martingale component is itself an additive functional. The first is a linear time trend, the second an additive martingale, the third a stationary process with mean zero, and the fourth a time-invariant constant. If we happen to impose the initialization: $Y_0=-\kappa_{+}\left(X_0\right)$, then the fourth term is zero. We use a Proposition 4.2.1 decomposition as a way to associate a "permanent shock" with an additive functional. The permanent shock is the increment to the martingale.
