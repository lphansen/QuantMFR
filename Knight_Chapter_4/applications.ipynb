# 4.3 Applications

We now compute martingale increments for two models of economic time series.

Application to a VAR
We apply the four-step construction in algorithm 4.2 when the Markov state $\left\{X_t\right\}$ follows a first-order VAR
$$
X_{t+1}=\mathbb{A} X_t+\mathbb{B} W_{t+1},
$$

where $\mathbb{A}$ is a stable matrix and $\left\{W_{t+1}: t \geq 0\right\}$ is a sequence of independent and identically normally distributed random variables with mean vector zero and identity covariance matrix. The one-step ahead conditional covariance matrix of the time $t+1$ shocks $B W_{t+1}$ to $X_{t+1}$ equals $B B^{\prime}$. Let
$$
Y_{t+1}-Y_t=\kappa\left(X_t, W_{t+1}\right)=\mathbb{D} X_t+\nu+\mathbb{F} W_{t+1},
$$
where $D$ and $F$ are row vectors with the same dimensions as $X_t$ and $W_{t+1}$, respectively, and the $(\cdot)$ symbol denotes an inner product. For this example, the four steps of algorithm 4.2 become:
(i) The trend growth rate is $\nu$ as specified.
(ii)
$$
\kappa_h\left(X_{t-1}, W_t, X_t\right)=\mathbb{D} X_{t-1}+\mathbb{F} W_t+\mathbb{D}(\mathbb{I}-\mathbb{A})^{-1} X_t
$$
(iii)
$$
\kappa_{+}(x)=\mathbb{D} x+\mathbb{D}(\mathbb{I}-\mathbb{A})^{-1} \mathbb{A} x
$$
(iv)
$$
\begin{aligned}
\kappa_m\left(X_{t-1}, W_t\right) & =\mathbb{F} W_t+\mathbb{D}(\mathbb{I}-\mathbb{A})^{-1}\left(X_t-\mathbb{A} X_{t-1}\right) \\
& =\left[\mathbb{F}+\mathbb{D}(\mathbb{I}-\mathbb{A})^{-1} \mathbb{B}\right] W_t
\end{aligned}
$$

From Example 1.10.1, we expect the coefficient of martingale increment to be the sum of impulse responses for the increment process $\left\{\mathbb{D} X_t+\mathbb{F} W_{t+1}\right.$ : $t \geq 0\}$. The impulse response function is the sequence of vectors:
$$
\mathbb{F}, \mathbb{D B}, \mathbb{D} \mathbb{A B}, \mathbb{D} \mathbb{A}^2 \mathbb{B}, \cdots \text {. }
$$

Summing these vectors gives
$$
\mathbb{F}+\mathbb{D}\left(\mathbb{I}+\mathbb{A}+\mathbb{A}^2+\cdots\right) \mathbb{B}=\mathbb{F}+\mathbb{D}(\mathbb{I}-\mathbb{A})^{-1} \mathbb{B}
$$
as anticipated.

Growth-Rate Regimes
We construct a Proposition 4.2 .1 decomposition for a model with persistent switches in the conditional mean and volatility of the growth rate $Y_{t+1}-Y_t$.

Suppose that $\left\{X_t: t \geq 0\right\}$ evolves according to an $n$-state Markov chain with transition matrix $\mathbb{P}$. Realized values of $X_t$ are coordinate vectors in $\mathbb{R}^n$. Suppose that $\mathbb{P}$ has only one unit eigenvalue. Let $\mathbf{q}$ be the row eigenvector associated with that unit eigenvalue normalized so that $\mathbf{q} \cdot \mathbf{1}_n=1$ and
$$
\mathbf{q}^{\prime} \mathbb{P}=\mathbf{q}^{\prime} .
$$

Consider an additive functional satisfying
$$
Y_{t+1}-Y_t=\mathbb{D} X_t+X_t^{\prime} \mathbb{F} W_{1, t+1},
$$
where $\left\{W_{1, t}\right\}$ is an i.i.d. sequence of multivariate standard normally distributed random vectors. Evidently, the stationary Markov $\left\{X_t: t \geq 0\right\}$ process induces discrete changes in both the conditional mean and the conditional volatility of the growth rate process $\left\{Y_{t+1}-Y_t\right\}$.
Observe that $E\left(X_{t+1} \mid X_t\right)=\mathbb{P} X_t$ and let
$$
W_{2, t+1}=X_{t+1}-E\left(X_{t+1} \mid X_t\right) .
$$

Thus we can represent the evolution of the Markov chain as
$$
X_{t+1}=\mathbb{P} X_t+W_{2, t+1}
$$
$\left\{W_{2, t+1}: t \geq 0\right\}$ is an $n \times 1$ discrete-valued vector process that satisfies $E\left(W_{2, t+1} \mid X_t\right)=0$, which is therefore a martingale increment sequence adapted to $X_t, X_{t-1}, \ldots, X_0$.
We again apply the four-step construction in algorithm 4.2.3
i)
$$
\nu=\mathbb{D} \mathbf{q}
$$
ii)
$$
H_t=\mathbb{D}\left(X_{t-1}-\mathbf{q}\right)+X_{t-1}^{\prime} \mathbb{F} W_{1, t}+\mathbb{D}\left((\mathbb{I}-\mathbb{P})^{-1} X_t\right.
$$

iii)
$$
H_t^{+}=E\left(H_{t+1} \mid X_t\right)=\mathbb{D}\left(X_t-\mathbf{q}\right)+\mathbb{D}(\mathbb{I}-\mathbb{P})^{-1} \mathbb{P} X_t
$$
which implies that
$$
\kappa_{+}(x)=\mathbb{D}\left(X_t-\mathbf{q}\right)+\mathbb{D}(\mathbb{I}-\mathbb{P})^{-1} \mathbb{P} x
$$
iv)
$$
G_t=H_t-H_{t-1}^{+}=X_{t-1}{ }^{\prime} \mathbb{F} W_{1, t}+\mathbb{D}(\mathbb{I}-\mathbb{P})^{-1} W_{2, t}
$$
where we have substituted from equation (4.6).
The martingale increment has both continuous and discrete components:
$$
\kappa_m\left(X_t, W_{t+1}\right)=\underbrace{X_t^{\prime} \mathbb{F} W_{1, t+1}}_{\text {continuous }}+\underbrace{\mathbb{D}(\mathbb{I}-\mathbb{P})^{-1}}_{\text {discrete }} W_{2, t+1} .
$$
