# 1. Discrete Framework

## 1.1 Functionals

Suppose $X_t$ is an $n$ dimensional, stationary, ergodic Markov process governed by the following relationship:

$$
X_{t+1} = \phi \left(X_t, W_{t+1}\right),\tag{1}
$$

Where $W_{t+1}$ is a $k$ dimensional i.i.d. shock process. 

```{margin} Example 1
A typical example of an additive functional is:

$$
Y_{t+1} - Y_t = \beta(X_t) + \alpha(X_t) \cdot W_{t+1}
$$

```

We call a process $Y$ an **additive functional** if it evolves as:

$$
Y_{t+1} - Y_t = \kappa \left(X_t, W_{t+1}\right),\tag{2}
$$

Where $Y_{t+1} - Y_t$ inherits the stationarity of $X_t$.

We are interested in macroeconomic processes whose *logarithms* behave as in equation 2. We can think of these as processes whose percentage change over time is stationary.

We might also be interested in the *levels* of these processes $\exp(Y)$, which we call **multiplicative functionals**, satisfying:

$$
\exp(Y_t+1)/\exp(Y_t) = \exp(\kappa \left(X_t, W_{t+1}\right)),\tag{3}
$$

In particular, we are interested in two cases of multiplicative functionals:
- $G$, which captures macroeconomic growth
- $S$, which captures stochastic discounting

<br>

## 1.2 Parameterizing Shock Exposure

<!-- Consider an asset whose payoff in each period evolves as a multiplicative functional:

$$
\log \left(G_t \right) - \log \left(G_{t-1}\right) = \beta \left(X_{t-1}\right) + \alpha \left(X_{t-1}\right) \cdot W_{t},\tag{4}
(\beta \left (X_{s} \right) + \alpha \left (X_{s} \right) \cdot W_{s+1})
$$ -->
Suppose we have an asset $G$ whose payoff in each period evolves as a multiplicative functional. We can write the log payoff at time $t$ as:

$$
\log \left(G_t \right) = \log \left(G_{0}\right) + \sum_{s=0}^{t-1} \kappa(X_t, W_{t+1}),\tag{5}
$$




<!-- $$
G_t / G_0 = \exp( \sum_{s=0}^{t-1} (\beta_g \left (X_{s}) \right + \alpha_g \left (X_{s}) \right \cdot W_{s+1}))
$$ -->

Suppose we were to increase the exposure of asset $G$ to some linear combination $\nu$ of the time 1 shocks $W_1$ by factor $r$, where the direction of exposure $\nu(X_0)$ is allowed to depend on the state $X_0$.

Then the time $t$ payoff becomes $G_t H_1(r)$, where:

$$
\begin{align}
\log H_1({\sf r}) = {\sf r}\nu(X_0)  W_1 - \frac {\sf r^2} 2 |\nu(X_0)|^2,\tag{7}
\end{align}
$$

We normalize the row vector $\nu$ so that ${\mathbb E} \left[|\nu(X_0)|^2\right] = 1.$
Notice that $H_1({\sf r})$ is positive and has conditional expectation equal to one.

Here we interpret $H_1(r)$ as a variable which shifts the exposure of $G_t$ to $W_1$, but we can also intepret $H_1(r)$ as a variable which changes the distribution of $W_1$ by giving it a conditional mean ${\sf r} \nu(X_0)$.


<!-- Then the log payoff at time $t$ becomes:

$$
\beta \left (X_{0} \right) + \left( \alpha \left (X_{0} \right) + r\nu \left (X_{0} \right) \right) \cdot W_{1} + \sum_{s=1}^{t-1} (\beta_g \left (X_{s} \right) + \alpha_g \left (X_{s} \right) \cdot W_{s+1}),\tag{6}
$$ -->

Thus, we obtain a family of expected payoffs parameterized by $r$:

$$
\begin{align}
{\mathbb E} \left[ \left( G_t \right) H_1({\sf r}) \mid X_0 \right],\tag{8}
\end{align}
$$


<!-- The first of these applications provides a distributional version of an impulse response function.  It matches exactly for the linear, log-normal model, in which case the $x$ plays no role. Once we include nonlinearities, the state $x$ can matter along with the time horizon $t$.  The second application explores the intertemporal asset pricing by determining the implied compensations for the changes in the exposures.  -->

<br>

## 1.3 Shock-Exposure Elasticities

We are interested in determining the percentage change in expected payoff ${\mathbb E} \left[ \left( G_t \right) H_1({\sf r}) \mid X_0 \right]$ corresponding to a small increase in $W_1$ in direction $\nu(X_0)$. Therefore, we want to compute the following **shock-exposure elasticity**:

$$
\begin{align}
\varepsilon_g(x,t) = \frac d {d{\sf r}}\log {\mathbb E} \left[ G_t H_1({\sf r}) \mid {X_0=x} \right]\vert_{{\sf r} = 0},\tag{9}
\end{align}
$$

We call this expression an elasticity, since we are differentiating a logarithm and our exposure direction $\alpha_h(X_0) \cdot W_1$ has unit standard deviation. We can rewrite equation 9 in a more tractable form by differentiating:

$$
\begin{align}
\varepsilon_g(x,t) = \frac {1}{{\mathbb E} \left[G_t H_1(r)\right]} {\mathbb E} \left[G_t \frac {d{H_1(r)}} {d{\sf r}} \mid {X_0=x} \right]\vert_{{\sf r} = 0} \\
= \frac {1}{{\mathbb E} \left[G_t H_1(r)\right]} {\mathbb E} \left[G_t \left( \nu(X_0) W_1 - r |\nu(X_0)|^2 \right) H_1(r) \mid {X_0=x} \right]\vert_{{\sf r} = 0} \\
= \frac {\nu(x) \cdot {\mathbb E} \left[ \left( G_t \right)  W_1 \mid X_0 = x\right]}{{\mathbb E} \left[ \left( G_t \right)\mid X_0 = x\right]},\tag{10}
\end{align}
$$

We can also calculate an equivalent elasticity for shocks at time $t$:

$$
\begin{align}
{\varepsilon}_g^+(x,t) = \frac d {d{\sf r}}\log {\mathbb E} \left[ \left(G_t \right)  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} =  \frac {{\mathbb E} \left[ \left( G_t \right) \nu(X_{t-1}) \cdot W_t \mid X_0 = x\right]}
{{\mathbb E} \left[ \left( G_t \right)\mid X_0 = x\right]},\tag{11}
\end{align}
$$

<!-- This calculation is degenerate in the case of linear log-normal specifications as it is constant over $t$.  
But it is sometimes of interest when nonlinearities are present.   -->

<!-- 
Our computations are local, performed by computing derivatives with respect to ${\sf r},$ and evaluating those derivatives at $\sf r =0:$

$$
\begin{align}
\varepsilon_M(x,t) = \frac d {d{\sf r}}\log {\mathbb E} \left[ \left(\frac { M_t}{M_0} \right) H_1({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} = \frac {\nu(x) \cdot {\mathbb E} \left[ \left( \frac {M_t}{M_0} \right)  W_1 \mid X_0 = x\right]}{{\mathbb E} \left[ \left( \frac {M_t}{M_0} \right)\mid X_0 = x\right]}
\end{align}
$$ -->

## 1.4 Shock-Price Elasticities

We are also interested in how risk compensations respond to shocks. Let $S_t$ be the period $t$ stochastic discount factor. Then the expected time $t$ return of an asset is given by its payoff divided by its price:

$$
R_t = \frac {{\mathbb E} [G_t]} {{\mathbb E} [S_t G_t]}
$$

Any increase (decrease) in an asset's risk exposure must be compensated by an increase (decrease) in its return. Hence, we can determine the 'price' of a shift in an asset's shock exposure $H_1(r)$ by evaluating how its return changes in response to a small increase in $r$, which gives us our definition for **shock-price elasticity**:

$$
\begin{align}
\varepsilon_{p}(x,t) = \frac d {d{\sf r}}\log {\mathbb E} \left[ \left( G_t \right) H_1({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0}
- \frac d {d{\sf r}}\log {\mathbb E} \left[ S_t G_t H_1({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} 
\end{align}
$$

Again, we can compute the shock-price elasticity for a change in $W_t$ exposure as:

$$
\begin{align}
{\varepsilon}_{p}^+(x,t) = \frac d {d{\sf r}}\log {\mathbb E} \left[ \left( G_t \right)  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0}
- \frac d {d{\sf r}}\log {\mathbb E} \left[ S_t G_t  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} . 
\end{align}
$$

Finally, consider how a change in shock exposure changes the value of an asset to an investor. This can be evaluated as the difference between the change in payoff and the change in risk premium corresponding to a small change in $r$. Thus the **value elasticity** is given by:

$$
\begin{align*}
& \frac d {d{\sf r}}\log {\mathbb E} \left[ S_t G_t  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} = {\varepsilon}_{g}^+(x,t) - {\varepsilon}_{p}^+(x,t) \\ 

{\varepsilon}_{p}^+(x,t) = & \frac d {d{\sf r}}\log {\mathbb E} \left[ \left( G_t \right)  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0}
- \frac d {d{\sf r}}\log {\mathbb E} \left[ S_t G_t  H_t({\sf r}) \mid X_0 \right]\vert_{{\sf r} = 0} . 
\end{align*}
$$



<!-- 
where

$$
\begin{align}
\frac {{\mathbb E} \left[ \left(G_t\right) H_1({\sf r}) \mid X_0 \right]}{{\mathbb E} \left[ \left( \frac {S_t G_t}{S_0 G_0} \right) H_1({\sf r}) \mid X_0 \right]}
\end{align}
$$

is the expected return on a payoff $G_t H_1({\sf r}),$  with cost

$$
\begin{align}
{\mathbb E} \left[ \left( \frac {S_t G_t}{S_0 G_0}\right) H_1({\sf r}) \mid X_0 \right].
\end{align}
$$ -->
